```python
target=df['Churn']
X=df.drop(['Churn'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,target,test_size=.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

logit=LogisticRegression()
logit.fit(X_train,y_train)
```

```python
logit.coef_
```

```python
logit.intercept_
```

```python
import statsmodels.api as sm
X_train_sm=sm.add_constant(X_train)
sm_model=sm.Logit(y_train,X_train_sm).fit()
```

```python
print(sm_model.summary())
```

```python
coef_sm=pd.read_html(sm_model.summary().tables[1].as_html(),header=0,index_col=0)[0]
```

```python
coef_sm['odds_ratio']=np.exp(coef_sm['coef'])
```

```python
coef_sm
```

```python
#predictions from statsmodel
X_test_sm=sm.add_constant(X_test)
pred_sm=sm_model.predict(X_test_sm)
```

```python
pred=np.where(pred_sm>0.5,1,0)
```

```python
#prdictions from sklearn model
y_pred=logit.predict(X_test)
```

```python
##confusion matrix of sm
from sklearn.metrics import confusion_matrix,precision_score,recall_score
print('statmodel')
print(confusion_matrix(y_test,pred))
print(confusion_matrix(y_test,y_pred))
```

```python
chart_data=X_test[cols]
```

```python
chart_data['prob']=prob[:,1]
```

```python
chart_data.head()
```

```python
chart_data.sort_values(by='prob',inplace=True,ascending=False)
```

```python
chart_data.head()
```

```python
chart_data.shape
```

```python
chart_data['decile']=pd.qcut(chart_data['prob'],10)
```

```python
chart_data.head()
```

```python
chart_data['row']=np.arange(1,chart_data.shape[0]+1)
```

```python
chart_data['decile1']=np.ceil(chart_data['row']/((chart_data.shape[0]+1)/10))
```

```python
chart_data.head()
```

```python
chart_data.decile1.value_counts()
```

```python
chart_data['target']=y_test
```

```python
lift=chart_data.groupby('decile1').agg(
    ['count' #totaal no. of customers in the decile
     ,'sum' # total no. of bad customers (risk=1)
    ])['target'].sort_index(ascending=True)
lift.columns=['No_customers','No_Left']
lift['Cumulative_left']=lift['No_Left'].cumsum()


#calculate gain= cumulative percent of events/bads

lift['Percent_of_events']=lift['No_Left']/lift['No_Left'].sum()*100
lift['Gain']=lift['Percent_of_events'].cumsum()

## calculate lift= ratio of bads to the no. of data points in the decile

lift['lift']=lift['Gain']/np.array(range(10,100+10,10))
lift.head(10)
```
